```python
import torchaudio as ta
from chatterbox.tts import ChatterboxTTS
```

```python
model = ChatterboxTTS.from_pretrained(device='cuda')
```

```python
from IPython.display import Audio
```

```python
wav =  model.generate(text, audio_prompt_path='enter path of the audio you want to clone',exaggeration = 0.4)
ta.save('enter path where output file will be saved', wav, model.sr)
Audio(filename='enter path where output file will be saved')
```
